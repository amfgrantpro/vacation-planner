# 🟢 Groq: What models are available right now?

Groq frequently adds, renames, and deprecates models. Run the two cells below
whenever you want an up-to-date list of what's currently live on the API.

````python
import os
from dotenv import load_dotenv
from groq import Groq

# Load environment variables from .env file
load_dotenv()

client = Groq(api_key=os.environ["GROQ_API_KEY"])
print("✅ Groq client ready")
````

````python
response = client.models.list()
models = sorted(response.data, key=lambda m: m.id)

print(f"{'MODEL ID':<45} {'OWNED BY':<15} {'ACTIVE':<8} {'CONTEXT WINDOW'}")
print("-" * 85)
for model in models:
    active = getattr(model, "active", "")
    context_window = getattr(model, "context_window", "")
    print(f"{model.id:<45} {model.owned_by:<15} {str(active):<8} {context_window}")

print(f"\nTotal models available: {len(models)}")
````
